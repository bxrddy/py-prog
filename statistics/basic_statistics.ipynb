```python
# import libraries

import pandas as pd
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('Baseball.csv')
df.head()
```

# Mean

```python
# mean of runs scored (RS)
df.RS.mean()
```

```python
# mean of runs scored (RS) by the arizona team (ARI)
df.loc[df.Team == "ARI"].RS.mean()
```

```python
# mean of runs scored (RS) by the arizona team (ARI) since 2005
df.loc[(df.Team == "ARI") & (df.Year > 2005)].RS.mean()
```

```python
# mean of runs allowed (RA) by the chicago team (CHC) until 2007
df.loc[(df.Team == "CHC") & (df.Year < 2007)].RA.mean()
```

# Median

```python
# median and mean of the wins (W)
df.W.median()
```

```python
# median of wins of the baltimore team (BAL) until 2000
df.loc[(df.Team == "BAL") & (df.Year <= 2000)].W.median()
```

# Mode

```python
# mode of OBP during the year 2010
df.loc[df.Year == 2010].OBP.mode()
```

#Correlation

```python
# pick variables
df_correlation = df[['RS', 'RA', 'W']]
df_correlation.head()
```

```python
# correlation matrix
df_correlation.corr()
```

```python
# correlation heatmap
sns.heatmap(data = df_correlation.corr(),
            annot = True,
            fmt = '.2g',
            center = 0,
            linewidth = 0.5,
            cmap = 'crest')
```

```python
# correlation between OBP, SLG, and BA
df_correlation1 = df[['OBP', 'SLG', 'BA']]
df_correlation1.head()
```

```python
df_correlation1.corr()
```

```python
sns.heatmap(data = df_correlation1.corr(),
            annot = True,
            fmt = '.2g',
            center = 0,
            linewidth = 0.7,
            cmap = 'rocket')
```
